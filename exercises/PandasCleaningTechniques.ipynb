## Missing Data

Try out different methods to locate and resolve missing data.

```python
# Try to find some missing data!

df.isna()
```

```python
df['Age'].isna()
```

```python
count_missing_ages = df['Age'].isna().sum()
print(count_missing_ages)
```

```python
count_missing_titles = df['Title'].isna().sum()
print(count_missing_titles)
```

```python
count_missing_text = df['Review Text'].isna().sum()
print(count_missing_text)
```

```python
cols = {'Title': 'No Title'}
df.fillna(value=cols)
```

```python
cols = {'Review Text': 'No Text Provided'}
df.fillna(value=cols)
```

```python
# this shows which columns contain missing data
missing_data = df.isnull().sum()
print(missing_data)
```
